```python
import pandas as pd
```

```python
loss_dat = pd.DataFrame(columns = ['epoch', 'loss', 'val_loss', 'acc', 'val_acc'])
counter = 1
with open("logs/logs_4conv.txt") as f:
    epo_num = ""
    for line in f:
        print(line)
        if counter % 2 != 0:
            epo_num = line.split(' ')[1].split('/')[0]
            loss_dat.loc[int(epo_num)-1, "epoch"] = int(epo_num)
        else:
            loss_dat.loc[int(epo_num)-1, 'loss'] = line.split(' ')[5]
            loss_dat.loc[int(epo_num)-1, 'acc'] = line.split(' ')[8]
            loss_dat.loc[int(epo_num)-1, 'val_loss'] = line.split(' ')[11]
            loss_dat.loc[int(epo_num)-1, 'val_acc'] = line.split(' ')[14]
        counter += 1
```

```python
loss_dat[['loss', 'val_loss', 'acc', 'val_acc']] = loss_dat[['loss', 'val_loss', 'acc', 'val_acc']].astype(float)
loss_dat
```

```python
lines = loss_dat[['acc', 'val_acc']].plot.line()
```

```python
fig = lines.get_figure()
fig.savefig('4conv_acc.png')
```
